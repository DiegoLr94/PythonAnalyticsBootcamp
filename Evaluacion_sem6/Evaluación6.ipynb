```python
df_south2=df_south.drop(columns=['region',
                                 'TEMPERATURA DO PONTO DE ORVALHO (°C)',
                                 'TEMPERATURA MÁXIMA NA HORA ANT. (AUT) (°C)',
                                'TEMPERATURA MÍNIMA NA HORA ANT. (AUT) (°C)', 
                                 'TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (°C)',
                                'TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (°C)',
                                'UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)',
                                'UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)',
                                'UMIDADE RELATIVA DO AR, HORARIA (%)',
                                'VENTO, DIREÇÃO HORARIA (gr) (° (gr))',
                                'VENTO, RAJADA MAXIMA (m/s)',
                                'VENTO, VELOCIDADE HORARIA (m/s)',
                                'station_code',
                                'latitude',
                                'longitude',
                                'height'])
df_south2
```

```python
profile = ppr.ProfileReport(df_south2, title='Pandas Profiling Report', minimal=True)
```

```python
profile.to_file('Reporte_inicial_data_south-Diego_Lopez.html')
```

```python
index_precipitacion=df_south2[df_south2['PRECIPITAÇÃO TOTAL, HORÁRIO (mm)']==-9999].index
index_precipitacion
```

```python
df_southa=df_south2.drop(index_precipitacion)
df_southa
```

```python
df_southa=df_south2.drop(index_precipitacion)
df_southa.reset_index()
```

```python
index_presionatm_estacion=df_southa[df_southa['PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)']==-9999].index
df_southb=df_southa.drop(index_presionatm_estacion)
df_southb.reset_index()
```

```python
index_presionmax=df_southb[df_southb['PRESSÃO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)']==-9999].index
df_southc=df_southb.drop(index_presionmax)
df_southc.reset_index()
```

```python
index_presionmin=df_southc[df_southc['PRESSÃO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)']==-9999].index
df_southd=df_southc.drop(index_presionmin)
df_southd.reset_index()
```

```python
index_radiacion=df_southd[df_southd['RADIACAO GLOBAL (Kj/m²)']==-9999].index
df_southe=df_southd.drop(index_radiacion)
df_southe.reset_index()
```

```python
index_temp=df_southe[df_southe['TEMPERATURA DO AR - BULBO SECO, HORARIA (°C)']==-9999].index
df_southf=df_southe.drop(index_temp)
df_southf.reset_index()
```

```python
df_southf['Date_dt']=pd.to_datetime(df_southf['Data'],format="%Y/%m/%d")
df_southf['Date_dt'].head()
```
